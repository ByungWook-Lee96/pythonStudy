```python
'''
1 부터 10까지의 정수 중에서 임의의 정수 10개를 
numbers 라는 list에 저장하기 
'''

numbers = []

count = 0
# fot number in range(10):
for number in range(1, 11):
    count += 1
    numbers.append(randint(1,10))
    print(count,'회차 : numbers :', numbers)

print('=' * 60)
print('numbers :', numbers)

# pirnt(random())
```

```python
'''
비어있는 list를 생성하고
1부터 10까지의 임의의 정수를 반지름으로 하는
원의 넓이 10개를 저장하는 list 를 출력하세요
'''
import random
list = []

for i in range(10):
    radius = randint(1, 10)
    list.append(radius * radius * 3.14)

print(list)
```

```python
'''
randint(start, stop) 함수
        ㄴ start 부터 stop까지 임의의 정수를 반환함

        start <= randint(start, stop) <= stop

random() 함수 : 0 부터 1 미만의 임의의 실수를 반환함

        0 <= random() < 1

randrange(start, stop) 함수
        ㄴ start 부터 stop - 1까지 임의의 정수를 반환함

        start <= randint(start, stop) < stop

'''

print(random.random())

# 1, 2, 3, 4, 5
print('randint(1, 5) :', randint(1, 5))

# 1, 2, 3, 4
print('randrange(1, 5) :', randrange(1, 5))
```

```python
'''
shuffle(리스트) 함수 : 리스트의 item 들을 임의로 섞어줌
    ㄴ 
'''
letters = ['가', '나', '다', '라', '마', '바']

# 반환값이 없어서 shuffle() 함수 자체를 출력하면 None이 출력됨
print('random.shuffle(letters) :', shuffle(letters))
print('letters :', letters)
```

```python
'''
choice(리스트) 함수 : 리스트에서 임의의 item을 반환함
'''

letters = ['가', '나', '다', '라', '마', '바']

print('choice(letters) :', choice(letters))

booleans = [True, False]
print('choice(booleans) :', choice(booleans))

dice_numbers = [1, 2, 3, 4, 5, 6]
print('choice(dice_numbers) :', choice(dice_numbers))
print('randint(1, 6) :', randint(1, 6))
print('randrange(1, 7) :', randrange(1, 7))
```

```python
'''
중첩 for 문 : nested for (loop) statement

구구단 - 
        (외부)  (내부)
        (고정)  (증가-반복)
        단수 X 곱하는 수
         2      (1 ~ 9)
         3      (1 ~ 9)
         4      (1 ~ 9)
         ...
         9      (1 ~ 9)
         
        
'''

# 2 단
print('2 X 1 = 2')
print('2 X 2 = 4')
print('2 X 3 = 6')
print('2 X 4 = 8')
print('2 X 5 = 10')
print('2 X 6 = 12')
print('2 X 7 = 14')
print('2 X 8 = 16')
print('2 X 9 = 18')


for number in range(1,10):
    print(f'2 X {number} = {2 * number}')
    
print('-' * 60)

for number in range(2,10):
    print(f'{number} 단')
    for change_num in range(1, 10):
        print(f'{number} X {change_num} = {number * change_num}')
```

```python
'''
짝수 단만 출력하세요
'''
for dan in range(2,10):  
    if dan % 2 == 0:
        print(dan, '단')
        for num in range(1,10):
            print(f'{dan} X {num} = {dan * num}')
```

```python
'''
홀수 단만 출력하세요
'''
for dan in range(2,10):  
    if dan % 2 != 0:
        print(dan, '단')
        for num in range(1,10):
            print(f'{dan} X {num} = {dan * num}')

    
```

```python
'''
각 단에 홀수만 곱하는 구구단을 출력하세요
'''

for dan in range(2, 10):
    for number in range(1, 10):
        if number % 2 == 1:
            print(f'{dan} X {number} = {dan * number}')
    print()
```

```python
'''
짝수만 곱하는 구구단을 출력하세요 : if not number % 2 -> number % 2가 0이면 False, not이라 반대 True 가 됨(실행) 
'''

for dan in range(2, 10):
    for number in range(1, 10):
        if not number % 2:
            print(f'{dan} X {number} = {dan * number}')
    print()
```

```python
'''
짝수단에는 짝수만 곱하고
홀수단에는 홀수만 곱하는
구구단을 출력하세요
'''
for dan in range(2,10):
    for num in range(1,10):
        if dan % 2 == 0:
            if num % 2 == 0:
                print(f'{dan} X {num} = {dan * num}')
        else:
            if num % 2 != 0: 
                print(f'{dan} X {num} = {dan * num}')
    print()
```

```python
for dan in range(2,10):
    for num in range(1,10):
        if dan % 2 == num % 2:
            print(f'{dan} X {num} = {dan * num}')
    print()
```

```python
str1 = '''나는 클라우드입니다
주소는 가산디지털단지입니다
나이는 23살 입니다'''

sentences = []
words = []

'''
str1의 문장에서
sentences 에는 문장단위로 저장하고
words에는 단어단위('띄어쓰기 기준')로 저장하고
문장의 개수와 단어의 개수를출력하세요 
'''

for sentence in str1.split('\n'):
    sentences.append(sentence)
    for word in sentence.split(" "):
        words.append(word)
        
print('문장의 개수 :', len(sentences), '개')
print('단어의 개수 :', len(words), '개')
print('sentences :', sentences)
print('words :', words)
```

```python
solutions = [1, 2, 3, 4, 5]
answers = [1, 3, 4, 2, 2]
'''
solutions은 정답지이고
answers는 학생이 기입한 답안입니다
이 두 list를 사용해서 
학생의 정답률을 출력하세요
'''
correct = 0
for i in range(len(solutions)):
    if solutions[i] == answers[i]:
        correct += 1
print(correct/len(solutions) * 100, '%')
```

```python
'''
주민등록번호를 입력받아서 유효성 검사를 하는 프로그램을 작성하세요(for문 사용하기)
    
    801212 - 2464874
    801212 - 246487
X   234567   892345

total : 각 자리수에 아래의 숫자를 곱한 숫자들의 합

11 - (total % 11) == 마지막 자리수 <-- 유효함
'''

jumin = input('주민번호를 입력하세요 : ').replace('-','')

correct = '234567892345'
total = 0

for i in range(len(correct)):
    total += int(jumin[i]) * int(correct[i])
print(total)

result = 11 - (total % 11)

if result >= 10:
    result %= 10
    
if result == int(jumin[-1]):
    print("유효한 값")
else:
    print("유효하지 않은 값")
```
