```python
'''
  클래스 와 객체(Object / instance)
  
  클래스 : 복잡한 형태의 data 를 메모리에 올리기 전에
           어떤 형태인지 설명해 놓은 것 (설명서) 
           
           클래스 = 속성   +    기능   
                    property    method
                    (멤버)변수  (멤버)메소드(함수)                    
           
  객체(Object / instance)
         :  클래스에 설명해 놓은대로 
            (실행 가능한 상태로) 메모리에 올린 data
            
           객체 = 속성 + 기능 
           
  객체를 생성하는 code <-- 클래스이름()         
           
  
  학생(Student) 클래스
  이름
  나이
  주소
  학년
  키
  몸무게
  
  name = '아마존'
  age  = 23
  address = '서울'
  grade = 3
  height = 187
  weight = 81  
  
'''
class Student:
    # 생성자 메소드 : 클래스의 멤버변수들을 초기화함
    #  ㄴ 객체를 생성할 때 자동으로 호출되는 메소드 
    def __init__(self):
        self.name = '아마존'
        self.age  = 23
        self.address = '서울'
        self.grade = 3
        self.height = 187
        self.weight = 81
        
    def display_info(self):
        print(f'''이름 : {self.name}
나이 : {self.age}
주소 : {self.address}
학년 : {self.grade}
키   : {self.height}
체중 : {self.weight}''')
    
'''
Student 객체를 생성하는 명령(code) <-- Student() 
  ㄴ 클래스에 설명한대로 data 를 메모리에 올리라는 명령
                          ㄴ 멤버변수
     생성자 메소드 : 객체를 생성할 때 자동으로 호출되는 메소드                          
'''
student1 = Student()   
print(student1.name)
print(student1.age)
print(student1.address)
print(student1.grade)
print(student1.height)
print(student1.weight)

'''
student1 이 저장하고 있는 주소지에 있는 
data 로 display_info() 메소드를 호출함(실행시킴)
'''
student1.display_info()
print('-' * 60)

student2 = Student()
print(student2.name)
print(student2.age)
print(student2.address)
print(student2.grade)
print(student2.height)
print(student2.weight)

student2.name = '이순신'
print('student1.name :',student1.name)
print('student2.name :',student2.name)

student1.name = '강감찬'
print('student1.name :',student1.name)
print('student2.name :',student2.name)
student2.age = 72
student2.address = '강릉'
student2.grade = 2
student2.height = 179
student2.weight = 75

'''
student2 가 저장하고 있는 주소지에 있는 
data 로 display_info() 메소드를 호출함(실행시킴)
'''
student2.display_info()
```

```python
print('-- student1 주소지의 정보로 display_info() 호출함 --')
student1.display_info()
print('-' * 60)

print('-- student2 주소지의 정보로 display_info() 호출함 --')
student2.display_info()
```

```python
class Person:
    
    # 생성자 메소드 : 클래스의 멤버변수들을 초기화함
    #  ㄴ 객체를 생성할 때 자동으로 호출되는 메소드
    def __init__(self, name, age, addr, grade):
        self.name = name
        self.age  = age
        self.addr = addr
        self.grade = grade
        
    def work(self):
        print(f'{self.name} 은(는) 열심히 일합니다')
    
    def gotowork(self):
        print(f'{self.name} 은(는) 직장에 출근합니다')
        
    def introduce(self):
        print(f'''이름 : {self.name}
나이 : {self.age}        
주소 : {self.addr}
학년 : {self.grade}''')
        
person1 = Person('장보고', 32, '청해진', 1)     
print('person1.name :',person1.name)
print('person1.age :',person1.age)
print('person1.addr :',person1.addr)
print('person1.grade :',person1.grade)

person2 = Person('안중근', 30, '황해도', 2)
print('person2.name :',person2.name)
print('person2.age :',person2.age)
print('person2.addr :',person2.addr)
print('person2.grade :',person2.grade)

print('-' * 60)
person1.work()
person1.gotowork()
person1.introduce()

print('-' * 60)
person2.work()
person2.gotowork()
person2.introduce()
```

```python
class Tv:
    def __init__(self, color, channel, power):
        self.color = color
        self.channel = channel
        self.power = power
    def power_on_off(self):
        self.power = not self.power
    def channel_down(self, number):
        self.channel -= number
    def channel_up(self, number):
        self.channel += number
        
tv1 = Tv('white', 1, False)
print('tv1.color :',tv1.color)
print('tv1.channel :',tv1.channel)
print('tv1.power :',tv1.power)   

print('-' * 60)

tv1.power_on_off()
print('tv1.power :',tv1.power)    

tv1.channel_up(10)
print('tv1.channel :',tv1.channel)

tv1.channel_down(5)
print('tv1.channel :',tv1.channel)

print('-' * 60)

'''
tv1 변수가 저장한 주소를
tv2 변수에 할당함 (복사-붙여넣기)
  ㄴ 객체를 하나 더 생성한 것이 아니고
     단지 주소만 복사함  <-- 주소가 같다는 것은 같은 객체라는 의미임
'''
tv2 = tv1
print('tv1.channel :',tv1.channel)
print('tv2.channel :',tv2.channel)
print('-' * 60)

tv2.channel_up(2)
print('tv2.channel :',tv2.channel)
print('tv1.channel :',tv1.channel)
print('-' * 60)

print('-- Tv 객체를 하나 더 생성해서 객체의 주소를 tv3 에 할당함 --')
tv3 = Tv('skyblue', 5, False)
print('tv3.color :',tv3.color)
print('tv3.channel :',tv3.channel)
print('tv3.power :',tv3.power)  
print('-' * 60)

tv3.channel_up(5)
print('tv3.channel :',tv3.channel)
print('tv1.channel :',tv1.channel)
print('tv2.channel :',tv2.channel)
```
