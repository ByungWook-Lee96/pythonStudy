```python
'''
 자료형    <--  data 를 관리하는 (메모리) 공간의 크기와 모양
 data type
 
 형변환
  ㄴ 자료형 -->  형
     data type   type
     
 int / float / bool / str   
  
'''

# int --> float : float() 함수 사용
num1 = 1234
print('num1 :',num1)
print('num1 :',type(num1))

num2 = float(num1)
print('num2 :',num2)
print('num2 :',type(num2))

# float --> int : int() 함수 사용
num3 = 123.456
print('num3 :',num3)
print('num3 :',type(num3))

num4 = int(num3)
print('num4 :',num4)
print('num4 :',type(num4))
```

```python
# int --> str : str()

num5 = 1234
print('num5 :',num5)
print('num5 :',type(num5))

result = num5 + 20
print('result :',result)

num6 = str(num5)
print('num6 :',num6)
print('num6 :',type(num6))

# TypeError: can only concatenate str (not "int") to str
# result = num6 + 20

result = num6 + str(20)
print('result :',result)
```

```python
# float  -->  str  : str()
num7 = 3.14
print('num7 :',num7)
print('num7 :',type(num7))

result = num7 + 20
print('result :',result)

num8 = str(num7)
print('num8 :',num8)
print('num8 :',type(num8))

# TypeError: can only concatenate str (not "int") to str
# result = num8 + 20

result = num8 + str(20)
print('result :',result)
```

```python
# str  -->  int  : int()
# 숫자모양의 문자열만 int type 으로 형변환 됨

# ValueError: invalid literal for int() with base 10: 'hello'
# num9 = 'hello'
num9 = '123456'
print('num9 :',num9)
print('num9 :',type(num9))

num10 = int(num9)
print('num10 :',num10)
print('num10 :',type(num10))
```

```python
# str  --> float  : float()
num11 = '3.141592'
print('num11 :',num11)
print('num11 :',type(num11))

num12 = float(num11)
print('num12 :',num12)
print('num12 :',type(num12))
```

```python
# 정수모양의 문자열  -->  float type
num13 = '123456'
print('num13 :',num13)
print('num13 :',type(num13))

num14 = float(num13)
print('num14 :',num14)
print('num14 :',type(num14))
```
