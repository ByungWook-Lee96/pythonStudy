```python
'''
list 복사하기
'''

list1 = [1, 2, 3, 4, 5]

# 주소의 복사 : 리스트를 복사해서 
#               그 복사한 리스트를 또 메모리에 올린 것이 아님
list2 = list1

print('list1 :',list1)
print('list2 :',list2)
print('list1 :',id(list1))
print('list2 :',id(list2))

print('list1[2] :',list1[2])
list1[2] = 3333
print('list1[2] :',list1[2])
print('list1 :',list1)

print('-' * 60)
print('list2[2] :',list2[2])
```

```python
'''
일차원 리스트에서의 복사
'''
list1 = [1, 2, 3, 4, 5]
print('list1 :',list1)

# print('list1[2:4] :',list1[2:4])

# 1) slicing 을 사용한 복사
list2 = list1[:]
print('list2 :',list2)

print('list1 :',id(list1))
print('list2 :',id(list2))

print('list1[4] :',list1[4])
list1[4] = 5555
print('list1[4] :',list1[4])
print('list1 :',list1)

print('-' * 60)
print('list2[4] :',list2[4])
print('list2 :',list2)
```

```python
# 2) copy() 메소드를 사용한 복사

list1 = [1, 2, 3, 4, 5]
print('list1 :',list1)

list2 = list1.copy()
print('list2 :',list2)

print('list1 :',id(list1))
print('list2 :',id(list2))
```

```python
'''
이차원 리스트의 경우

1) 얕은 복사
'''

list1 = [[1, 2, 3],[4, 5, 6]]

list2 = list1[:]
list2 = list1.copy()

print('list1 :', list1)
print('list2 :', list2)
print('list1 :', id(list1))
print('list2 :', id(list2))
```

```python
print('list1[0] :',list1[0])
print('list2[0] :',list2[0])
print('list1[0] :',id(list1[0]))
print('list2[0] :',id(list2[0]))

print('list1[1] :',list1[1])
print('list2[1] :',list2[1])
print('list1[1] :',id(list1[1]))
print('list2[1] :',id(list2[1]))
```

```python
list1 = [[1, 2, 3],[4, 5, 6]]

list2 = list1[:]
list2 = list1.copy()

print('list1 :', list1)
print('list2 :', list2)

print('list1[0][1] :',list1[0][1])
print('list2[0][1] :',list2[0][1])

list1[0][1] = 2222
print('list1[0][1] = 2222')

'''
얕은 복사 :  일차원 리스트는 복사되지만, 
             이차원 리스트는 복사되지 않음 
             (두 개의 일차원 리스트가 이차원 리스트를 공유함) 

list1[0][1] = 2222 <-- list1 의 [0][1] 번째 item 을 2222 로 변경함
list1[0][1] = 2222 <-- list2 의 [0][1] 번째 item 도 2222 로 변경됨
'''

print('list1[0][1] :',list1[0][1])
print('list2[0][1] :',list2[0][1])

print('list1 :', list1)
print('list2 :', list2)
```

```python
'''
이차원 리스트의 경우

2) 깊은 복사
    ㄴ copy 모듈에 있는 deepcopy() 함수를 사용함
            ㄴ import 해서 사용하라는 의미
            
            형식]
            
            copy.deepcopy(리스트)
'''
import copy

list3 = [[1, 2, 3],[4, 5, 6]]
list4 = copy.deepcopy(list3)

print('list3 :', list3)
print('list4 :', list4)
print('list3 :', id(list3))
print('list4 :', id(list4))
```

```python
print('list3[0] :',list3[0])
print('list4[0] :',list4[0])
print('list3[0] :',id(list3[0]))
print('list4[0] :',id(list4[0]))

print('list3[1] :',list3[1])
print('list4[1] :',list4[1])
print('list3[1] :',id(list3[1]))
print('list4[1] :',id(list4[1]))
```

```python
list3[0][1] = 2222
print('list3[0][1] = 2222')

'''
깊은 복사 : 일차원 리스트도 복사되고, 이차원 리스트로 복사됨

list3[0][1] <-- list3 의 [0][1] 번째 item 을 2222 로 변경함
list3[0][1] <-- list4 의 [0][1] 번째 item 은 영향 받지 않음
'''

print('list3[0][1] :',list3[0][1])
print('list4[0][1] :',list4[0][1])


print('list3 :', list3)
print('list4 :', list4)
```
