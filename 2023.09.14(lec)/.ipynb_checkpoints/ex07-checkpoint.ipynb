```python
'''
정수 세 개를 입력 받아서  <-- input()
최댓값과 최솟값을 출력하세요
max      min
'''

num1 = int(input('첫 번째 숫자를 입력하세요 : '))
num2 = int(input('두 번째 숫자를 입력하세요 : '))
num3 = int(input('세 번째 숫자를 입력하세요 : '))

'''
num1 이 최댓값이 되는 조건
num1 >= num2 and num1 >= num3   <-- max = num1

num2 가 최댓값이 되는 조건
num2 >= num1 and num2 >= num3

num3 가 최댓값이 되는 조건
num3 >= num1 and num3 >= num2
'''
max = 0

if num1 >= num2 and num1 >= num3:
    max = num1
elif num2 >= num1 and num2 >= num3:
    max = num2
elif num3 >= num1 and num3 >= num2:
    max = num3    
    
print(f'최댓값 : {max}')
```

```python
num1 = int(input('첫 번째 숫자를 입력하세요 : '))
num2 = int(input('두 번째 숫자를 입력하세요 : '))
num3 = int(input('세 번째 숫자를 입력하세요 : '))

max = 0
min = 0

if num1 >= num2 and num1 >= num3:
    max = num1
elif num2 >= num3:
    max = num2
else:
    max = num3    
    
print(f'최댓값 : {max}')

if num1 <= num2 and num1 <= num3:
    min = num1
elif num2 <= num3:
    min = num2
else:
    min = num3  
    
print(f'최솟값 : {min}')
    
```

```python
'''
정수 두 개를 입력 받아서  <-- input()
최댓값과 최솟값을 출력하세요
max      min
'''

num1 = int(input('첫 번째 숫자를 입력하세요 : '))
num2 = int(input('두 번째 숫자를 입력하세요 : '))

max = 0
min = 0

if num1 >= num2:
    max = num1
else:
    max = num2
    
if num1 <= num2:
    min = num1
else:
    min = num2   
    
print(f'최댓값 : {max}')  
print(f'최솟값 : {min}')
```

```python
'''
if 문을 사용한 삼항연산자

형식]

결과  =  참일때값 if 조건문 else 거짓일때값
'''

num1 = int(input('첫 번째 숫자를 입력하세요 : '))
num2 = int(input('두 번째 숫자를 입력하세요 : '))
   
max = num1 if num1 >= num2 else num2
min = num1 if num1 <= num2 else num2

print(f'최댓값 : {max}')  
print(f'최솟값 : {min}')  
```

```python
'''
정수 세 개를 입력 받아서  <-- input()
최댓값과 최솟값을 출력하세요
max      min
'''

num1 = int(input('첫 번째 숫자를 입력하세요 : '))
num2 = int(input('두 번째 숫자를 입력하세요 : '))
num3 = int(input('세 번째 숫자를 입력하세요 : '))

max  =  num1 if num1 >= num2 and num1 >= num3 else num2 if num2 >= num3 else num3
min  =  num1 if num1 <= num2 and num1 <= num3 else num2 if num2 <= num3 else num3

print(f'최댓값 : {max}')  
print(f'최솟값 : {min}') 
```

```python
'''
번호   통신사
011    SKT
016    KT
019    LGT

번호에 따라서 통신사를 출력하는
code 를 작성하세요
'''

phone = input('전화번호를 입력하세요 : ')

# 0116598574 : 입력한 번호에서 앞 세 자리만 slicing 함
phone = phone[:3]

company = ''

if phone == '011':
    company = 'SKT'
elif phone == '016':
    company = 'KT'    
elif phone == '019':
    company = 'LGT' 
else:
    company = '?'
    
if company != '?':    
    print(f'통신사는 {company} 입니다')  
else:
    print('전화번호를 올바르게 입력해 주세요')
```
